```python
#Open the text file for processing:
file = open("mytext.txt","r") 
```

```python
text=file.read()
```

```python
text
```

```python
#import libraries
from nltk import sent_tokenize
from nltk import word_tokenize
```

```python
#Sentence tokenizing
sentences=sent_tokenize(text)
```

```python
sentences
```

```python
#words tokenizing
words=word_tokenize(text)
```

```python
words
```

```python
#Find the frequency distribution:
from nltk.probability import FreqDist
fdist=FreqDist(words)
```

```python
fdist
```

```python
#print 10 most common words
fdist.most_common(10)
```

```python
#Plot the frequency graph:
import matplotlib.pyplot as plt
fdist.plot(10)
```

```python
#Remove punctuation marks:
#empty list to store the values
words_no_punc = []

#removing punctuation marks:
for w in words:
    if w.isalpha():
        words_no_punc.append(w.lower())
print(words_no_punc)  
print('\n')
print (len(words_no_punc))   
    

```

```python
#Plotting graph without punctuation marks:
fdist=FreqDist(words_no_punc)
fdist.plot(10)
```

```python
#Open the text file for stowords:
stopwords = open("stop.txt","r") 
stop_words=stopwords.read()
print(stop_words)
```

```python
#removing stopwords

#empty list to store clean words
clean_words = []
for w in words_no_punc:
    if w not in stop_words:
        clean_words.append(w)
print(clean_words)        
```

```python
#Final Frquency distribution
fdist=FreqDist(clean_words)
fdist.plot(10)
```

```python
#stemming Example
from nltk.stem import PorterStemmer 
porter = PorterStemmer()
word_list=["liked,likes,likey,liking"]
for w in word_list:
    print(porter.stem(w))
```

```python
#lemmatizing:It gives dictionary word
from nltk.stem import WordNetLemmatizer   
lemmatizer = WordNetLemmatizer()
print("better :", lemmatizer.lemmatize("better", pos ="a")) 
```
